```python
%matplotlib inline
import numpy as np
from sympy import fwht
import matplotlib.pyplot as plt
```

```python
def custom_fwht(x):
    n = len(x)
    if n == 1:
        return x
    first_half = custom_fwht(x[:n // 2])
    x_even = first_half + custom_fwht(x[n // 2:])
    x_odd = first_half - custom_fwht(x[n // 2:])
    return np.concatenate((x_even, x_odd))
```

## Source function

```python
N = 32
x = np.arange(0, 2 * np.pi, 2 * np.pi / N)
y = np.array([np.sin(5 * t) + np.cos(2 * t) for t in x])
plt.plot(x, y)
```

## Custom Fast Walsh transform

```python
plt.plot(x, custom_fwht(y))
```

## Library Fast Walsh transform

```python
plt.plot(x, fwht(y))
```
